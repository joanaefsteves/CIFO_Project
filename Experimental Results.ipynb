```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Read the Grid Search results
results1 = pd.read_csv('fitness_per_run.csv')
results2 = pd.read_csv('avg_fitness_per_generation.csv')
```

```python
results2
```

```python
results2["elitism_flag"] = np.where(
    results2["elitism"] != 0, "elitism_true", "elitism_false"
)
```

```python
results2["combination"] = (
    results2["mutation"] + " + " +
    results2["crossover"] + " + " +
    results2["selection"] + " + " +
    results2["elitism_flag"]
)
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=results2, x="generation", y="avg_fitness", hue="combination", marker="o")
plt.title("Evolution of Average Fitness per Generation")
plt.xlabel("Generation")
plt.ylabel("Average Fitness")
plt.grid(True)
plt.tight_layout()
plt.show()
```
